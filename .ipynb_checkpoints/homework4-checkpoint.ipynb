```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import data

```python
file = "data/TSLA.csv"

df = pd.read_csv(file, decimal = ".")

df = df.iloc[:, [0, 4]]

df.head(5)
```

```python
df["Date"] = pd.to_datetime(df["Date"], format = "%Y-%m-%d") #konvertiert Date Collumn damit es als datum erkannt wird
```

```python
df.dtypes
```

```python
df_2022 = df[(df["Date"]>="2022-01-03") & (df["Date"]<"2022-12-30")]
df_2022.shape
                                           
```

```python
y = df_2022["Close"]
x = df_2022["Date"]

plt.plot(x, y, label = "Closing price")
plt.xticks(rotation=45)
plt.title("Tesla, Inc. (TSLA)")
plt.legend()
plt.show()
```
